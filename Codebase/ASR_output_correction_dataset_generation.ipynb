```python
import pandas as pd
import os

def open_csv_files(file_paths):
  """Opens and processes CSV files specified in a list of file paths.
  Args:
    file_paths: A list of strings, where each string is a valid path to a CSV file.
  """
  for file_path in file_paths:
    if not os.path.exists(file_path):
      print(f"Error: File not found - {file_path}")
      continue
    try:
      # Attempt to open the file with pandas
      df = pd.read_csv(file_path, sep='\t')  # Use appropriate separator
      print(f"Successfully opened: {file_path}")
      # print(df.head()) # Print first few rows for verification
    except pd.errors.ParserError as e:
        print(f"Error parsing {file_path}: {e}")
    except Exception as e:
      print(f"An error occurred while processing {file_path}: {e}")

file_list = [
    "/content/drive/MyDrive/datasets/common-voice-19/cv-corpus-19.0-delta-2024-09-13/en/validated.tsv",
    "/content/drive/MyDrive/datasets/common-voice-20/cv-corpus-20.0-delta-2024-12-06/en/validated.tsv",
    "/content/drive/MyDrive/datasets/common-voice-18/cv-corpus-18.0-delta-2024-06-14/en/validated.tsv",
    "/content/drive/MyDrive/datasets/common-voice-17/cv-corpus-17.0-delta-2024-03-15/en/validated.tsv"
]
open_csv_files(file_list)
```

```python
import os
import pandas as pd

def process_and_combine_csv(file_paths, parent_paths):
    """Processes multiple CSV files, keeps specific columns, and combines them.

    Args:
        file_paths: A list of file paths to CSV files.
        parent_paths: A list of parent directories for recordings to prepend to 'path'.

    Returns:
        A pandas DataFrame containing the combined data, or None if an error occurs.
    """
    all_dataframes = []
    for file_path, parent_path in zip(file_paths, parent_paths):
        if not os.path.exists(file_path):
            print(f"Error: File not found - {file_path}")
            continue
        try:
            df = pd.read_csv(file_path, sep='\t')
            if 'sentence' in df.columns and 'path' in df.columns:
                df = df[['sentence', 'path']]
                df['path'] = parent_path + "/" + df['path']
                all_dataframes.append(df)
            else:
                print(f"Warning: 'sentence' or 'path' column not found in {file_path}. Skipping.")
        except pd.errors.ParserError as e:
            print(f"Error parsing {file_path}: {e}")
            continue
        except Exception as e:
            print(f"An error occurred while processing {file_path}: {e}")
            continue

    if not all_dataframes:
        print("No valid dataframes to combine.")
        return None

    return pd.concat(all_dataframes, ignore_index=True)

file_list = [
    "/content/drive/MyDrive/datasets/common-voice-19/cv-corpus-19.0-delta-2024-09-13/en/validated.tsv",
    "/content/drive/MyDrive/datasets/common-voice-20/cv-corpus-20.0-delta-2024-12-06/en/validated.tsv"
    "/content/drive/MyDrive/datasets/common-voice-18/cv-corpus-18.0-delta-2024-06-14/en/validated.tsv",
    "/content/drive/MyDrive/datasets/common-voice-17/cv-corpus-17.0-delta-2024-03-15/en/validated.tsv"
]
parent_paths = [
    "/content/drive/MyDrive/datasets/common-voice-19/cv-corpus-19.0-delta-2024-09-13/en/clips",
    "/content/drive/MyDrive/datasets/common-voice-20/cv-corpus-20.0-delta-2024-12-06/en/clips",
    "/content/drive/MyDrive/datasets/common-voice-18/cv-corpus-18.0-delta-2024-06-14/en/clips",
    "/content/drive/MyDrive/datasets/common-voice-17/cv-corpus-17.0-delta-2024-03-15/en/clips"
]

data = process_and_combine_csv(file_list, parent_paths)
# print(len(data))
if data is not None:
    print(data['path'][0])
```

## Loading the audio [generating time-domain representations with a specified sampling rate]

```python
def load_audio_from_dataframe(df, sr=16000, duration=None, offset=0.0):
  """Loads audio files from a dataframe.
  Args:
    df: Pandas DataFrame with 'path' column containing audio file paths.
    sr: Sampling rate for audio loading.
    duration: Duration of audio to load (in seconds).
    offset: Starting offset for audio loading (in seconds).

  Returns:
    A list of audio time series, or None if an error occurs.
  """
  c = 0
  audio_data = []
  for index, row in df.iterrows():
    try:
        #print(file_path)
        y, sr = librosa.load(row['path'], sr=sr, duration=duration, offset=offset, mono=True)
        audio_data.append([y, row['sentence'], row['path']])
    except FileNotFoundError:
        c +=1
        # print(f"File not found: {row['path']}")
    except Exception as e:
        print(f"Error loading audio file {row['path']}: {e}")
        return None
  print(f"Number of files not found: {c}")
  return audio_data

audio_timeseries = load_audio_from_dataframe(data, sr=16000, duration=5) # Load first 5 seconds of each audio file

print(f"Loaded {len(audio_timeseries)} audio files.")
audio_timeseries = pd.DataFrame(audio_timeseries, columns=['audio', 'sentence', 'path'])
```

## Generating data with ASR models

```python
# cache for models for efficient usage
models = dict({})
processors = dict({})
```

### whisper-tiny

```python
if models.get("whisper-tiny") is None:
  processors["whisper-tiny"] = AutoProcessor.from_pretrained("openai/whisper-tiny")
  models["whisper-tiny"] = AutoModelForSpeechSeq2Seq.from_pretrained("openai/whisper-tiny")
```

### facebook/s2t-small-librespeech

```python
processors["s2t-small-librispeech-asr"] = AutoProcessor.from_pretrained("facebook/s2t-small-librispeech-asr")
models["s2t-small-librispeech-asr"] = AutoModelForSpeechSeq2Seq.from_pretrained("facebook/s2t-small-librispeech-asr")
```

### *generate_transcripts* function [Generates the transcripts for audio file with ASR model specified in the arguments]

```python
def generate_transcripts(df, model_name):
    if model_name not in models:
        print(f"Error: Model '{model_name}' not found in the 'models' dictionary.")
        return None

    transcripts = []
    model = models[model_name]
    processor = processors[model_name]
    batch_size = 200

    for index, row in df.iterrows():
        try:
            audio = row['audio']
            input_features = processor(audio, sampling_rate=16000, return_tensors="pt").input_features.to(model.device)
            predicted_ids = model.generate(input_features)
            predicted_transcript = processor.batch_decode(predicted_ids, skip_special_tokens=True)[0]
            transcripts.append((row['path'], predicted_transcript, row['sentence']))
        except Exception as e:
            print(f"Error generating transcript for {row['path']}: {e}")
            # transcripts.append((row['path'], "Error", row['sentence']))

        if len(transcripts) >= batch_size:
            df_transcripts = pd.DataFrame(transcripts, columns=['audio_path', 'predicted_transcript', 'actual_sentence'])
            print("Saving batch of 200 transcripts...")
            df_transcripts.to_csv(f"t_batch_{index//batch_size}.csv", index=False)
            transcripts = []

    if transcripts:
        df_transcripts = pd.DataFrame(transcripts, columns=['audio_path', 'predicted_transcript', 'actual_sentence'])
        df_transcripts.to_csv(f"t_last_batch.csv", index=False)

    return df_transcripts
```

```python
t = generate_transcripts(audio_timeseries, "s2t-small-librispeech-asr")

import pandas as pd
df_t = pd.DataFrame(t, columns=['audio_path', 'predicted_transcript', 'actual_sentence'])

# Save the DataFrame to a CSV file in Google Drive
df_t.to_csv('/content/drive/MyDrive/datasets/t.csv', index=False)
```

## Cleaning the data based on the following constraints:
1. Total number of mismatched words in actual sentence and generated transcripts must not exceed 2.
2. Rows with transcripts other than english language should be removed.

```python
def clean_dataframe(df_t):
    removed_rows = []
    cleaned_df = []
    c=0
    for index, row in df_t.iterrows():
        predicted_words = row['predicted_transcript'].split()
        actual_words = row['actual_sentence'].split()
        word_diff = abs(len(predicted_words) - len(actual_words))

        if word_diff > 2:
            removed_rows.append(row)
            c+=1
            continue

        # Check if sentences contain only English characters
        if not re.fullmatch(r'[a-zA-Z\s.,!?;:\'\"]*', row['predicted_transcript']) or not re.fullmatch(r'[a-zA-Z\s.,!?;:\'\"]*', row['actual_sentence']):
            removed_rows.append(row)
            c+=1
            continue

        if row['predicted_transcript'] == row['actual_sentence']:
            removed_rows.append(row)
            c+=1
            continue

        cleaned_df.append(row)

    cleaned_df = pd.DataFrame(cleaned_df)

    print("Removed rows:",c)
    for row in removed_rows:
      print(row['predicted_transcript'], row['actual_sentence'])

    return cleaned_df

df_t = pd.read_csv('/content/asr_correction_dataset.csv') # Load your dataframe
cleaned_df_t = clean_dataframe(df_t)
cleaned_df_t.to_csv('/content/drive/MyDrive/datasets/asr_correction_dataset_cleaned.csv', index=False)
# print("\nCleaned dataframe:")
# print(cleaned_df_t.head())
```
